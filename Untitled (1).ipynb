Data Science Projects on Area and Population

```python
import pandas as pd
cities = pd.read_csv("california_cities.csv")
cities
```

```python
#extracting data that is needed
latitude, longitude = cities['latd'], cities['longd']
population, area = cities['population_total'], ['area_total_km2']
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn
seaborn.set()
plt.scatter(longitude, latitude, label=None, c=np.log10(population),
            cmap='viridis', s=area, linewidth=0, alpha=0.5)
plt.axis(aspect='equal')
plt.xlabel('Longitude')
plt.ylabel('Longitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)
# Create an empty list to hold the dummy scatter plots
dummy_plots = []

# Iterate over the area sizes
for area in [100, 300, 500]:
    # Create a dummy scatter plot with no data points, only for the legend
    dummy_plot = plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + " kmÂ²")
    dummy_plots.append(dummy_plot)

# Show the legend with the dummy plots
plt.legend(handles=dummy_plots, scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')

# Set the title of the plot
plt.title("Area and population of California cities")

# Show the plot
plt.show()
```
